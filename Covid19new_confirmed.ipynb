```python
import pandas as pd

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
df = pd.read_csv(url, error_bad_lines=False)
```

```python
conf_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv',error_bad_lines=False)
deaths_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv', error_bad_lines=False)
recv_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv',error_bad_lines=False)
```

```python
conf_df =conf_df.set_index('Country/Region')
deaths_df = deaths_df.set_index('Country/Region')
recv_df = recv_df.set_index('Country/Region')

conf_df.drop(index='Mainland China', inplace=True)
deaths_df.drop(index='Mainland China', inplace=True)
recv_df.drop(index='Mainland China', inplace=True)
```

```python
conf=conf_df.sum()
deaths= deaths_df.sum()
recv= recv_df.sum()
```

```python
conf = conf.drop(labels=['Lat', 'Long'])
deaths = deaths.drop(labels=['Lat', 'Long'])
recv = recv.drop(labels=['Lat', 'Long'])
```

```python
C=[]
C.append(conf[0])
for i in range(len(conf)-1):
    dif = conf[i+1] - conf[i]
    C.append(dif)
D=[]
D.append(deaths[0])
for i in range(len(deaths)-1):
    dif = deaths[i+1] - deaths[i]
    D.append(dif)
R=[]
R.append(recv[0])
for i in range(len(recv)-1):
    dif = recv[i+1] - recv[i]
    R.append(dif)
    
```

```python
dfsum = pd.DataFrame(
    {'Επιβεβαιωμένα κρούσματα': C,
     'Έχουν αναρρώσει': D,
     'Death': R
    })
```

```python
L = conf.index
```

```python
dfsum.set_index(L)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
dfsum.plot(kind='area', 
             alpha=0.25, # 0-1, default value a= 0.5
             stacked=False,
             figsize=(22, 13),
            )

plt.title(' Κρούσματα Covid-19')
plt.ylabel('Πληθος ατόμων')
plt.xlabel('Ημέρες')

plt.show()
```
